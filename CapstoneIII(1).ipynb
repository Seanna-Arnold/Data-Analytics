```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import missingno
import json
```

```python
#For this report we're going to see if international trade is linked to countries' economy
```

```python
ExportsByIncome_df = pd.read_csv("ExportsByIncome.csv")
ExportsByIncome_df.head()
```

```python
ExportsByIncome_df.drop_duplicates()
ExportsByIncome_df.head()# cleaning up the dataset
```

```python
missingno.matrix(ExportsByIncome_df, figsize = (30,10))
```

```python
#get rid of missing data for accuracy
```

```python
ExportsByIncome_df.drop('Flag Codes',axis=1, inplace=True)
ExportsByIncome_df.drop('Flags',axis=1, inplace=True)
```

```python
missingno.matrix(ExportsByIncome_df, figsize = (30,10))
```

```python
ExportsByIncome_df.plot(x='Country', y='Value', kind='line', figsize=(20,3), color='orange')
plt.show()
#We are going to use orange for the visualizations representing this dataframe
#This visualization shows the range of values in each level of income countries. We can see that some countries earn more than others on their level but not by much from their exports.
#From this visualization, we can see that the higher income countries have more value in the stuff they export
```

```python
ExportsByIncome_df.hist(column='Value', bins=3, grid=False, rwidth=9,
        color='orange', sharex=True) #here is overall value range/estimates concerning value from exports
```

```python
ExportsByIncome_df['Value'].plot(kind='box', figsize=(25, 6), color='orange')
plt.show()
#This shows the mean mode of lower and upper percentiles of export values
```

```python
ExportsByIncome_df['Value'].plot(kind='box', figsize=(25, 6))
plt.show()
```

```python
make = ExportsByIncome_df['Country']
horsepower = ExportsByIncome_df['Value']
 
# Figure Size
fig = plt.figure(figsize =(25, 20))
 
# Horizontal Bar Plot
plt.bar(make, horsepower, color='orange')
plt.xlabel('Countries', fontsize=14)
plt.ylabel('Value', fontsize=14)

# Show Plot
plt.show()
# In this visualization, you can see that the higher the economy, the higher its value from exports
```

```python
Trade_df = pd.read_csv("TradeBalanceByIncome.csv")
Trade_df.head()
```

```python
Trade_df.drop_duplicates()
Trade_df.head()
```

```python
missingno.matrix(Trade_df, figsize = (30,10))
```

```python
Trade_df.drop('Flag Codes',axis=1, inplace=True)
Trade_df.drop('Flags',axis=1, inplace=True)
```

```python
missingno.matrix(Trade_df, figsize = (30,10))
```

```python
Trade_df.plot(x='Country', y='Value', kind='line', figsize=(20,3), color='red')
plt.show()
# In this viusalization, we can see that Upper Middle income countries have the most value in trade and not the High income. So this part of trade doeas not directly effect economies
```

```python
Trade_df.hist(column='Value', bins=3, grid=False, rwidth=9,
        color='red', sharex=True)#here is overall value range/estimates concerning value from trade
```

```python
Trade_df['Value'].plot(kind='box', figsize=(25, 6), color='red')
plt.show() #This shows the mean mode of lower and upper percentiles of trade values
```

```python
countries = Trade_df['Country']
values = Trade_df['Value']
 
# Figure Size
fig = plt.figure(figsize =(25, 20))
 
# Horizontal Bar Plot
plt.bar(countries, values, color='red')
plt.xlabel('Countries', fontsize=14)
plt.ylabel('Values', fontsize=14)

# Show Plot
plt.show()
# In this visualization, you can see that the higher the economy, the higher its value from trade so we can conclude that international trade has great effect on controlling a country's economy
```
